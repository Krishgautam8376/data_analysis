```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("line Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid('true')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.bar(x, y)
plt.title("bar Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid('true')
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [1, 2, 3, 4, 5, 6, 7]
plt.hist(data)
plt.title("hist Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
data = [2, 3, 4, 5, 6]
labels = ['a', 'b', 'c', 'd', 'e']
plt.pie(data, labels=labels, autopct= '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.barh(x, y)
plt.title(" horizontal line Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```
